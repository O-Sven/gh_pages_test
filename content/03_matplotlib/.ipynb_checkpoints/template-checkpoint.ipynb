Task:

* read temperature data for a day
* construct a list, containing the time, and another one containign the temperatures
* plot the temperature vs. time

```python
from datetime import datetime

data_file = open('weather_data/16_11_2018_temperature.log', 'r')

time_list = []
temperature_list = []

time_refernce_str = '00:00:00'
time_refernce = datetime.strptime(time_refernce_str, '%H:%M:%S')

for raw_line in data_file:
    line = raw_line.strip().split(',')
    current_time = datetime.strptime(line[1], '%H:%M:%S')
    time_difference = current_time - time_refernce
    time_list.append(time_difference.total_seconds()/3600)
    temperature_list.append(float(line[2]))


#print(time_list)
plt.plot(time_list, temperature_list)
plt.show()
```

## Functions

### Task

* write a function which gets a file name and returns two lists: time [in hours] and temperature


```python
def read_weather_log(filename):
    data_file = open(filename, 'r')

    time_list = []
    temperature_list = []

    time_refernce_str = '00:00:00'
    time_refernce = datetime.strptime(time_refernce_str, '%H:%M:%S')

    for raw_line in data_file:
        line = raw_line.strip().split(',')
        current_time = datetime.strptime(line[1], '%H:%M:%S')
        time_difference = current_time - time_refernce
        time_list.append(time_difference.total_seconds()/3600)
        temperature_list.append(float(line[2]))
        
    return time_list, temperature_list
```

Task

* use above function to redraw the data
* add axis labels
* plot the data of three days in a single plot, add a legend

```python
fig = plt.figure(figsize=(14,7))

day_list = [14, 15, 16, 17, 18, 19]

for day in day_list:
    filename = 'weather_data/{:02d}_11_2018_temperature.log'.format(day)
    time, temp = read_weather_log(filename)
    
    plt.plot(time, temp, label='{}. November 2018'.format(day))

plt.legend()
plt.xlim(0,24)
plt.show()
```

Task

* manually adjust the x value range to [0, 24]
* change the x-ticks to show only [0,6,12,18,24] hours

```python
fig = plt.figure(figsize=(14,7))

day_list = [15, 16, 17, 18]

for day in day_list:
    filename = 'weather_data/{:02d}_11_2018_temperature.log'.format(day)
    time, temp = read_weather_log(filename)
    
    plt.plot(time, temp, label='{}. November 2018'.format(day))

plt.legend()

tick_values = [0,6,12,18,24]
tick_labels = ["{:d}".format(v) for v in tick_values]
plt.xticks(tick_values, tick_labels)
plt.xlabel('time / h')
plt.ylabel('air temparture / deg C')
plt.xlim(0,24)
plt.show()
```
